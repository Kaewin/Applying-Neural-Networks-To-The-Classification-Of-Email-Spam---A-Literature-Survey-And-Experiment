```python
import pandas as pd
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
# import data
d1 = pd.read_csv('data/enron_proccessed.csv')
d2 = pd.read_csv('data/enron_proccessed_spam.csv')
d3 = pd.read_csv('data/spamassassin_processed_1.csv')
d4 = pd.read_csv('data/spamassassin_processed_2.csv')
```

```python
# combine d1 and d2, update the index
data = pd.concat([d1, d2], ignore_index=True)
```

```python
# combine d1, d2, and d3. update the index
data = pd.concat([data, d3], ignore_index=True)
```

```python
# combine d1, d2, and d3. update the index
data = pd.concat([data, d4], ignore_index=True)
```

```python
data.tail()
```

```python
# print the columns in data
print(data.columns)
```

```python
# drop Message Body column
data.drop('Message Body', axis=1, inplace=True)

# drop the index column
data.drop('Index', axis=1, inplace=True)

# drop the index column
data.drop('Cleaned Body', axis=1, inplace=True)
```

```python
data.head()
```

```python
# Splitting the data into features and target variable
X = data.drop('Target', axis=1)  # Replace 'target_variable_name' with the actual name of your target variable
y = data['Target']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)  # Adjust the test_size and random_state as desired
```

```python
# check out X_train and X_test
print(X_train.shape)
print(X_test.shape)

# Also check out y_train and y_test
print(y_train.shape)
print(y_test.shape)
```

```python
# import the precision metric from tensorflow
from tensorflow.keras.metrics import Precision
```

```python
# Create the model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(7,)),
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['Precision'])

# Define the validation data
validation_data = (X_test, y_test)

# Add early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=100, verbose=1)

# Train the model
model.fit(X_train, y_train, epochs=10000, validation_data=validation_data, callbacks=[early_stopping])
```

```python
# Evaluate the model
model.evaluate(X_test, y_test)
```

```python
import io
import sys
import nltk
from nltk.corpus import stopwords

def get_email_text():
    """Prompts the user to input the copied raw text of an email."""

    email_text = ""

    print("Please copy and paste the raw text of an email:")

    while True:
        line = input()

        if line == "":
            break

        email_text += line + "\n"

    return email_text

def extract_features(email):
    features = {
        'number_of_words': len(email.split()),
        'number_of_stop_words': len([word for word in email.split() if word in list(stopwords.words('english'))]),
        'number_of_unique_words': len(set(email.split())),
        'ratio_of_lowercase_to_uppercase': float(len([word for word in email.split() if word.islower()])) / len(email.split()),
        'number_of_exclamation_points': email.count('!'),
    }

    # Tokenize the message body
    tokens = nltk.word_tokenize(email)

    # Remove stop words
    stop_words = nltk.corpus.stopwords.words('english')
    tokens = [token for token in tokens if token not in stop_words]

    # Stem the tokens
    stemmer = nltk.stem.PorterStemmer()
    stemmed_tokens = [stemmer.stem(token) for token in tokens]

    # Lemmatize the tokens
    lemmatizer = nltk.stem.WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

    features['number_of_unique_stemmed_words'] = len(set(stemmed_tokens))
    features['number_of_lemmatized_words'] = len(set(lemmatized_tokens))

    cleaned_body = ' '.join([stemmer.stem(token) for token in email.split() if token not in stop_words])

    return features
```

```python
from nltk.corpus import stopwords

email_text = get_email_text()

features = extract_features(email_text)

print("The features of the email are:")
print(features)
```
